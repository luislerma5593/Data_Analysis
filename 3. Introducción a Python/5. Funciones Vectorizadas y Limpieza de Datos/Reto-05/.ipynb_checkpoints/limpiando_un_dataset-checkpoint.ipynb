```python
import pandas as pd

df = pd.read_csv('C:/Users/LLerma/Google Drive/Analisis de Datos/M3 - Introduccion a Python/Files//melbourne_housing-raw.csv', sep=',')

df
```

```python
df.isna().sum()
```

```python
df = df.dropna(axis = 0, how = "any")
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
names = {'Suburb' : 'Suburb', 
         'Address': 'Address', 
         'Rooms': 'Rooms', 
         'Type': 'Type', 
         'Price': 'Price', 
         'Method': 'Method', 
         'SellerG': 'Seller',
         'Date': 'Date', 
         'Distance': 'Distance', 
         'Postcode': 'Postcode', 
         'Bedroom2': 'Bedroom_2', 
         'Bathroom': 'Bathroom', 
         'Car': 'Car',
         'Landsize': 'Landsize', 
         'BuildingArea': 'Building_Area', 
         'YearBuilt': 'Year_Built', 
         'CouncilArea': 'Council_Area', 
         'Lattitude': 'Lattitude',
       'Longtitude': 'Longtitude', 
         'Regionname': 'Region_Name', 
         'Propertycount': 'Property_Count'}
```

```python
df = df.rename(columns = names)
```

```python
df
```

```python
print("     PRICE ")
print("Min",df["Price"].min())
print("Max",df["Price"].max())
print("Mean",df["Price"].mean())
print("\n")
print("     DISTANCE ")
print("Min",  df["Distance"].min())
print("Max",df["Distance"].max())
print("Mean",df["Distance"].mean())
print("\n")
print("     LAND_SIZE ")
print("Min",df["Landsize"].min())
print("Max",df["Landsize"].max())
print("Mean",df["Landsize"].mean())
```
