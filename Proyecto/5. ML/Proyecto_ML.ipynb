# Filtrado $\mu \pm 3\sigma$

```python
h1.shape
```

```python
h2.shape
```

### Energía

```python
h1_energy_mean = h1["EnergyPerHeat"].mean()
h2_energy_mean = h2["EnergyPerHeat"].mean()

h1_energy_std = h1["EnergyPerHeat"].std()
h2_energy_std = h2["EnergyPerHeat"].std()
```

```python
h1_energy_filter = (h1["EnergyPerHeat"] > (h1_energy_mean - 3 * h1_energy_std)) & (h1["EnergyPerHeat"] < (h1_energy_mean + 3 * h1_energy_std))
h2_energy_filter = (h2["EnergyPerHeat"] > (h2_energy_mean - 3 * h2_energy_std)) & (h2["EnergyPerHeat"] < (h2_energy_mean + 3 * h2_energy_std))
```

---

### Power On

```python
h1_POn_mean = h1["PowerOn"].mean()
h2_POn_mean = h2["PowerOn"].mean()

h1_POn_std = h1["PowerOn"].std()
h2_POn_std = h2["PowerOn"].std()
```

```python
h1_POn_filter = (h1["PowerOn"] > (h1_POn_mean - 3 * h1_POn_std)) & (h1["PowerOn"] < (h1_POn_mean + 3 * h1_POn_std))
h2_POn_filter = (h2["PowerOn"] > (h2_POn_mean - 3 * h2_POn_std)) & (h2["PowerOn"] < (h2_POn_mean + 3 * h2_POn_std))
```

---

### Tap to Tap

```python
h1_TT_mean = h1["TapToTap"].mean()
h2_TT_mean = h2["TapToTap"].mean()

h1_TT_std = h1["TapToTap"].std()
h2_TT_std = h2["TapToTap"].std()
```

```python
h1_TT_filter = (h1["TapToTap"] > (h1_TT_mean - 50)) & (h1["TapToTap"] < (h1_TT_mean + 50))
h2_TT_filter = (h2["TapToTap"] > (h2_TT_mean - 50)) & (h2["TapToTap"] < (h2_TT_mean + 50))
```

---

### kWh / ton

```python
h1_kWh_mean = h1["EnergyPerChargeTon"].mean()
h2_kWh_mean = h2["EnergyPerChargeTon"].mean()

h1_kWh_std = h1["EnergyPerChargeTon"].std()
h2_kWh_std = h2["EnergyPerChargeTon"].std()
```

```python
h1_kWh_filter = (h1["EnergyPerChargeTon"] > (h1_kWh_mean - 100)) & (h1["EnergyPerChargeTon"] < (h1_kWh_mean + 100))
h2_kWh_filter = (h2["EnergyPerChargeTon"] > (h2_kWh_mean - 100)) & (h2["EnergyPerChargeTon"] < (h2_kWh_mean + 100))
```

---

### Aplicación de filtros

```python
h1_fil = h1[h1_energy_filter & h1_POn_filter & h1_kWh_filter & h1_TT_filter]
print("\tHORNO 1")
print(" Sin filtro:", len(h1))
print(" Sin filtro:", len(h1_fil))
```

```python
h2_fil = h2[h2_energy_filter & h2_POn_filter & h2_kWh_filter & h2_TT_filter]
print("\tHORNO 2")
print(" Sin filtro:", len(h1))
print(" Sin filtro:", len(h1_fil))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.4)
```

# Comparación entre los diferentes modelos de ML supervisados

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics

def desempeño(y_test,y_pred):
    matriz = confusion_matrix(y_test,y_pred)
    print(matriz)
    #(TN, FP, FN, TP) = resultado.ravel()
    print("\nExactitud:    ",round(metrics.accuracy_score(y_test, y_pred),6)*100,"%")
    print("Precision:    ",round(metrics.precision_score(y_test, y_pred),6)*100,"%")
    print("Sensibilidad: ",round(metrics.recall_score(y_test, y_pred),6)*100,"%")  
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

NaiveBayes = GaussianNB()
NaiveBayes.fit(x_train, y_train) 
y_pred_nb = NaiveBayes.predict(x_test) 

desempeño(y_test, y_pred_nb)
```

### Máquina de soporte vectorial

```python
from sklearn.svm import SVC #Support Vector Classifier

svc_m = SVC()
svc_m.fit(x_train,y_train)
y_pred_svc_m = svc_m.predict(x_test)
desempeño(y_test, y_pred_svc_m)
```

### Redes neuronales artificiales

```python
from sklearn.neural_network import MLPClassifier

nn_m = MLPClassifier(hidden_layer_sizes=(20,10,5),
                     max_iter=100000,
                     activation="logistic",
                     tol=1e-8)

nn_m.fit(x_train,y_train)
y_pred_nn_m = nn_m.predict(x_test)
desempeño(y_test, y_pred_nn_m)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

for_m = RandomForestClassifier(n_estimators=800)
for_m.fit(x_train,y_train)
y_pred_for_m = for_m.predict(x_test)
desempeño(y_test, y_pred_for_m)
```
