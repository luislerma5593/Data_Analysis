# Filtrado $\mu \pm 3\sigma$

```python
h1.shape
```

```python
h2.shape
```

### Energía

```python
h1_energy_mean = h1["EnergyPerHeat"].mean()
h2_energy_mean = h2["EnergyPerHeat"].mean()

h1_energy_std = h1["EnergyPerHeat"].std()
h2_energy_std = h2["EnergyPerHeat"].std()
```

```python
h1_energy_filter = (h1["EnergyPerHeat"] > (h1_energy_mean - 3 * h1_energy_std)) & (h1["EnergyPerHeat"] < (h1_energy_mean + 3 * h1_energy_std))
h1_energy_filter
```

```python
h2_energy_filter = (h2["EnergyPerHeat"] > (h2_energy_mean - 3 * h2_energy_std)) & (h2["EnergyPerHeat"] < (h2_energy_mean + 3 * h2_energy_std))
h2_energy_filter
```

### Power On

```python
h1_POn_mean = h1["PowerOn"].mean()
h2_POn_mean = h2["PowerOn"].mean()

h1_POn_std = h1["PowerOn"].std()
h2_POn_std = h2["PowerOn"].std()
```

```python
h1_POn_filter = (h1["PowerOn"] > (h1_POn_mean - 3 * h1_POn_std)) & (h1["PowerOn"] < (h1_POn_mean + 3 * h1_POn_std))
h1_POn_filter
```

```python
h2_POn_filter = (h2["PowerOn"] > (h2_POn_mean - 3 * h2_POn_std)) & (h2["PowerOn"] < (h2_POn_mean + 3 * h2_POn_std))
h2_POn_filter
```

---

### kWh / ton

```python
h1_kWh_mean = h1["EnergyPerChargeTon"].mean()
h2_kWh_mean = h2["EnergyPerChargeTon"].mean()

h1_kWh_std = h1["EnergyPerChargeTon"].std()
h2_kWh_std = h2["EnergyPerChargeTon"].std()
```

```python
h1_kWh_filter = (h1["EnergyPerChargeTon"] > (h1_kWh_mean - 100)) & (h1["EnergyPerChargeTon"] < (h1_kWh_mean + 100))
h1_kWh_filter
```

```python
h2_kWh_filter = (h2["EnergyPerChargeTon"] > (h2_kWh_mean - 100)) & (h2["EnergyPerChargeTon"] < (h2_kWh_mean + 100))
h2_kWh_filter
```

---

### Aplicación de filtros

```python
len(h1)
```

```python
h1_fil = h1[h1_energy_filter & h1_POn_filter & h1_kWh_filter]
len(h1_fil)
```

```python
len(h2)
```

```python
h2_fil = h2[h2_energy_filter & h2_POn_filter & h2_kWh_filter]
len(h2_fil)
```
